```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
fullData = pd.read_csv('FIFA_data.csv', index_col=0)
droppable_cols = ['Real Face', 'Joined', 'Contract Valid Until','Photo', 'Flag', 'Club Logo', 'Special', 'Body Type']
usefulData = fullData.drop(columns=droppable_cols, errors='ignore')
#usefulData.head()
```

```python
usefulData['TotalPassing'] = (usefulData['ShortPassing'] + usefulData['LongPassing'])/2
usefulData['GrowthPotential'] = usefulData['Potential'] - usefulData['Overall']
```

```python
def convert(value):
    #value = row.Value
    lastChar = value[-1]
    try:
        if lastChar == 'K':
            value = str(float(value[0:-1])*100)
        elif lastChar == 'M':
            value = str(float(value[0:-1])*100000)
    except: 
        if lastChar == 'K':
            value = str(float(value[0])*100)
        elif lastChar == 'M':
            value = str(float(value[0])*100000)
    return value
# end of def convert
```

```python
def symbolFix(value):
    'removes the symbols from the string'
    value = value.replace('€','')
    return value
```

```python
def convert_v2(value):
    'hopefully a better version of convert'
    value = symbolFix(value)
    
    lastChar = value[-1]
    if lastChar == 'K':
        factor = 1000
    elif lastChar == 'M':
        factor = 1000000
    else:
        return 0
        #print('OOOPS -', value)
        
    try:
        num = float(value[:-1])
    except:
        num = value[:-1]
        print(value)
        
    return num*factor
    
# end of def convert_v2
```

```python
usefulData.Value = usefulData.Value.apply(convert_v2)
```
